### Questions: 
1. Which genres are the most common?
2. Which genres have a high avg budget and revenue?
3. Which genres have a high avg profit?
4. Which genres have a high avg popularity?
5. Which genres have the highest number of movies along with a vote avergae >= 8

```python
movies['profit'] = movies['revenue'] - movies['budget']
```

```python
movies_genre = movies[['popularity', 'budget','revenue','original_title','runtime','genres','release_date','vote_count','vote_average','profit']]
```

```python
movies_genre.head()
```

```python
from pandas import Series, DataFrame
movies_genre['genres'].str.split('|').apply(Series,1).stack()
```

```python
split = movies_genre['genres'].str.split('|').apply(Series,1).stack()
split.index = split.index.droplevel(-1)
split.name = 'genres_split'
del movies_genre['genres']
movies_genre = movies_genre.join(split)
```

```python
movies_genre
```

## Question 1:
Which genres are the most common?

```python
genres_count = pd.DataFrame(movies_genre.groupby('genres_split').original_title.nunique()).sort_values('original_title',ascending = True)
```

```python
genres_count
```

```python
genres_count['original_title'].plot.pie(title = 'Movies Per Genre in %', autopct= '%1.1f%%', figsize = (10,10))
```

```python
genres_count['original_title'].plot.barh(title = 'Movies Per Genre', figsize = (10,10))
```

## Question 2:
Which genres have a high avg budget and revenue?

```python
genres_avg = movies_genre.groupby('genres_split').mean(numeric_only = True)
pd.options.display.float_format = '{:2f}'.format
genres_avg
```

```python
genres_avg.sort_values('budget',ascending = True, inplace = True)
```

```python
genres_avg[['budget','revenue']].plot.barh(title = 'Budget and Revenue by Genre', color = ('DarkBlue','LightBlue'),figsize = (10,10)).legend(loc='lower right')
```

## Question 3:
Which genres have a high avg profit?

```python
genres_avg.sort_values('profit',ascending = True, inplace = True)
```

```python
genres_avg['profit'].plot.barh(title = 'Profit by Genre',figsize = (10,10)).legend(loc='lower right')
```

## Question 4:
Which genres have a high avg popularity?

```python
genres_avg.sort_values('popularity',ascending = True, inplace = True)
```

```python
genres_avg['popularity'].plot.barh(title = 'Popularity by Genre',figsize = (10,10)).legend(loc='lower right')
```

## Look at Profit Per Genre Per Year

```python
movies_genre = movies[['popularity', 'budget','revenue','original_title','runtime','genres','release_year','vote_count','vote_average','profit']]
split = movies_genre['genres'].str.split('|').apply(Series,1).stack()
split.index = split.index.droplevel(-1)
split.name = 'genres_split'
del movies_genre['genres']
movies_genre = movies_genre.join(split)
movies_genre
```

```python
time_genre = pd.DataFrame(movies_genre.groupby(['release_year','genres_split'])['profit'].mean())
```

```python
time_genre
```

```python
profit_time = pd.pivot_table(time_genre,values = 'profit', index = ['genres_split'], columns = ['release_year'])
```

```python
sns.set(rc = {'figure.figsize':(15,10)})
sns.heatmap(profit_time, cmap = 'YlGnBu',linewidths = .5)
plt.title('Genres by Profit per Year')
```
